```python
# we'll need functions from the `math` module
import math

def e2w(energy):
  """
  converts energy to wavelength `energy`: energy in eV (float or int) returns wavelength in meters
  """

  h = 6.626070040E-34
  c = 2.99792458E8
  e = 1.6021766208E-19

  return (h * c) / (energy * e)

def float_range(start, stop, step):
  """ generate a range of float numbers"""

  # first get the whole range:
  r = stop - start

  # figure out how many numbers we'll have in the sequence given the step.
  # We use the `math.ceil` function to "round up", e.g., if the division results
  # something like 78.3, we will expect a total of 78 numbers, ensuring
  # we don't go beyong the "stop" value

  count = math.floor(r / step)

  # create the empty result list:
  result = []

  # now iterate over all the index values. The `range` function returns an
  # "iterator" with `count` numbers, starting from 0. To include also the
  # `number` itself, we need to count one further.
  # see https://www.w3schools.com/python/ref_func_range.asp
  for i in range(count + 1):
    result.append(start + step * i)

  return result

print(float_range(0.5, 20, 0.25))

def generate_table(start=0.5, stop=20, step=0.25):
  """ generate a list of calculated energy and wavelength pairs ("tuples") """

  # the `range` function is a useful Python function to generate a sequence
  # with a start, stop, and step.
  # Note that the `stop` parameter is NOT included by default
  r = float_range(start, stop, step)

  # we can print the range for debugging here:
  # print(r)

  # initialise empty list for results:
  result = []

  for energy_value in r:
    wavelength = e2w(energy_value)
    # store the result as a 2-dimensional "tuple"
    # (see https://www.w3schools.com/python/python_tuples.asp)
    # and then append it to the result list
    result.append((energy_value, wavelength))

  return result

def print_table(tuple_list, output=None):
  """ this will print out a given list of pairs as a table on the screen.
      an optional parameter `output` is used to make this function
      more generic. If `output` is `None` then it prints on the screen,
      but it can also be set to a file identified, in which case it is
      written to the file instead.
  """

  # One way to format output is using the `%` operator
  print('% 20s % 20s' % ('energy (eV)', 'wavelength (nm)'), file=output)
  for tuple_pair in tuple_list:
    # more powerful is the `format` method used on strings.
    # see https://realpython.com/python-formatted-output/
    # we multiply by 1e9 to convert from m to nm
    formatted_row = '{0:20.4f} {1:20.4f}'.format(tuple_pair[0],tuple_pair[1] * 1e9)
    print(formatted_row, file=output)

def write_txt_file(file_name, tuple_list):

  # we create a file to write (`w`) into
  with open(file_name, mode='w') as f:
    print_table(tuple_list, f)

def write_csv_file(file_name, tuple_list):
  """ this will write given list of pairs as a table into a file.
      We format this as a CSV file (https://en.wikipedia.org/wiki/Comma-separated_values)
      that can be read e.g. in Excel and other software
  """

  # we create a file to write (`w`) into
  with open(file_name, mode='w') as f:

    # instead of printing on screen, we now write into a file.
    # in a CSV file the first row is always the "headers"
    print('%s,%s' % ('energy (eV)', 'wavelength (nm)'), file=f)

    # same as above when printing on screen, but writing into a file now.
    # and using the `e` scientific formatting
    for tuple_pair in tuple_list:
      formatted_row = '{0:e},{1:e}'.format(tuple_pair[0],tuple_pair[1]*1e9)
      print(formatted_row, file=f)



```

```python
from math import pi
import matplotlib.pyplot as plt

def estimate_pi(N=10000, a=2, norm=6):
  # initialisiere die Variable `sum` als 0.0
  x = 0.0

  # Iteriere über eine range, die bei 1 started und bis N (inklusive) geht
  for n in range(1,N+1):
    # direkte Umsetzung der Formel im Summenzeichen, `**` ist "hoch", also "n hoch a"
    # `x += ...` ist das Gleiche wie `x = x + ...`
    x += 1 / (n**a)

  # Umstellen der Gleichung nach Pi
  pi_est = (x * norm) ** (1/a)

  return pi_est

K=10000
pi_2 = estimate_pi(N=K, a=2, norm=6)
pi_4 = estimate_pi(N=K, a=4, norm=90)
err_2 = abs(pi_2-pi)
err_4 = abs(pi_4-pi)

print('estimate with a=2: {0:1.10f}, error: {1:1.10f}, error %: {2:.8f}'.format(pi_2, err_2, 100 * err_2 / pi_2))
print('estimate with a=4: {0:1.10f}, error: {1:1.10f}, error %: {2:.8f}'.format(pi_4, err_4, 100 * err_4 / pi_4))

x_list = []
y_list = []
y2_list = []
for N in [1, 10, 50, 100, 500, 750, 1000, 5000, 10000, 50000]:
  pi_2 = estimate_pi(N=N, a=2, norm=6)
  pi_4 = estimate_pi(N=N, a=4, norm=90)
  err_2 = abs(pi_2-pi)
  err_4 = abs(pi_4-pi)
  x_list.append(N)
  y_list.append(err_2)
  y2_list.append(err_4)

plt.plot(x_list, y_list, label="a=2")
plt.plot(x_list, y2_list, label="a=4")
#plt.plot()

plt.xlabel("N")
plt.ylabel("Error")
plt.yscale('log')
plt.title("Pi Estimation")
plt.legend()
plt.show()

```

```python

def get_epsilon(dtype):
    eps = dtype(1)
    # loop while the sum of 1+epsilon is different to 1 using the respective type.
    # once it is the same, we have found the machine precision, it's the last
    # epsilon for which it was different, so we return that
    while dtype(1) + dtype(eps) != dtype(1):
        previous_eps = eps
        # half eps every loop (divide by 2)
        eps = dtype(eps) / dtype(2)
    return previous_eps

print('machine precision of type {0} is {1}'.format('int', get_epsilon(int)))
print('machine precision of type {0} is {1}'.format('float', get_epsilon(float)))
```
